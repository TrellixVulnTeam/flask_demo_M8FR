```python
#SH
import yfinance as yf #Allows us to Use Yahoo Finance API 
import pandas as pd
import numpy as np
```

```python
#SH
#Function to Get Different Kinds Of Info From a Stock Ticker Directly Through API 

def ticker_info():
    ticker_name = input("Enter Ticker Name")
    ticker_symbol = yf.Ticker(ticker_name)
    ticker_data = ticker_symbol.history(period="max")
    return ticker_data
```

```python
#SH
ticker_info()
```

```python
#SH
#Function to Pull Current Price from a Stock Ticker Directly Through API

def current_ticker_price():
    ticker_name = input("Enter Ticker Name")
    ticker_symbol = yf.Ticker(ticker_name)
    ticker_data = ticker_symbol.history(period="min")
    ticker_open = ticker_data.filter(['Open'])
    ticker_open = ticker_open.rename(columns = {"Open": ticker_name}) 
    print(ticker_open)
```

# BUILDING SUMMARIES BASED ON DATA

```python
#SH
#Building Summary for Stocks Remaining

total_stocks_remaining_summary = price_df_last_100.filter(regex="Stock").sum()
stocks_remaining_df = total_stocks_remaining_summary.where(total_stocks_remaining_summary > 0).to_frame() #Filtering Out Negative Values

stocks_remaining_df = stocks_remaining_df.rename(columns = {0: "Stocks Remaining"}) 
```

```python
#SH
#Sorting the Data By Column
price_df_sorted = price_df.reindex(sorted(price_df.columns), axis=1)

#Changing Row to Column
current_day_price_column = price_df_sorted[-1:].transpose()

#NEED THIS TO RELABEL WITH CONSTANTLY CHANGING COLUMN LABEL DUE TO API LIVE PULLING DATA
current_day_price_column.rename(columns = {list(current_day_price_column)[0]: 'Current Day Price'}, inplace = True)
```

```python
#SH
stocks_remaining_df["Current Day Price"] = current_day_price_column["Current Day Price"].values
stocks_remaining_df["Value of Stocks Owned"] = stocks_remaining_df["Stocks Remaining"] * stocks_remaining_df["Current Day Price"]

#Removing All NaNs
stocks_remaining_dropna = stocks_remaining_df.dropna()

total_stocks_remaining_summary = stocks_remaining_dropna["Value of Stocks Owned"].sum()
print("Total Value of Stocks Owned:",total_stocks_remaining_summary)
stocks_remaining_dropna
```

```python
#SH
total_earned_column_names = np.asarray(list(price_df_last_100.filter(regex="TOTAL NET")))

total_earned_summary = price_df_last_100[total_earned_column_names].sum()

print("Total Summary of User's Stock Gains/Losses:")

print(total_earned_summary)

print("Total Gains and Losses:", total_earned_summary.sum())
```

```python
#SH
#Final Net Portfolio Value

print("Final Net Portfolio Value:", total_stocks_remaining_summary + total_earned_summary.sum())
```
